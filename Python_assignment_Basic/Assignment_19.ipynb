Then, define the class Robot that has one instance (object) of each of these. Define a does() method for the Robot that prints what its component objects do.**

```python
class laser:
  def does():
    return  'disintegrate'
class claw:
  def does():
    return 'crush'
class smartphone:
  def does():
    return 'ring'
class robot:
  def __init__(self):
    self.laser = laser.does()
    self.claw = claw.does()
    self.smartphone = smartphone.does()
  def does(self):
    print (self.laser,':laser',self.claw,':claw',self.smartphone,':smart phone')
```
