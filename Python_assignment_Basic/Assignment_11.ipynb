considered the same, and &#39;goodbye&#39; and &#39;GOODbye&#39; are also considered the same)**


```python
def raise_assert(eggs,bacon):
  eggs = eggs.upper()
  bacon = bacon.upper()
  assert eggs.lower() != bacon.lower()



```

```python
raise_assert( 'hello' , 'hello')
```

```python
raise_assert('goodbye','GOODbye')
```

 **9. What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?**


**ANS:**

Step Over:

Functionality: Executes the current line of code and moves to the next line in the same function. If the current line contains a function  without stepping through its individual lines.

Step In:

Functionality: Moves the debugger into the next line of code. If the current line contains a function call, it enters that function and pauses at its first line.

Step Out:

Functionality: Executes the remaining lines of the current function and pauses when control returns to the calling function. It effectively completes the current function and steps back to the caller.
