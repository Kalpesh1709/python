**5. What is transferred to the keyword arguments allowRegexes and blockRegexes?**


 Ans:

 allowRegexes and blockRegexes keyword arguments are typically used  where you need to filter or validate strings based on regular expressions (regex).

```python
pip install pyinputplus
```

```python
import pyinputplus as pyip
response = pyip.inputStr(allowRegexes=['a-z'])
```

```python
response = pyip.inputStr(blockRegexes=[r'[2345]$'])
```
