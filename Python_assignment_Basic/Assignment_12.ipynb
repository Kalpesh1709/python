**5. What methods do you use to rotate a page?**

Ans:

rotateClockwise() and rotatecountertClockwise() this function inside module use for rotate page

```python
from PyPDF2 import PdfReader
from PyPDF2 import PdfWriter

with open('Tiger.pdf','+rb') as file:
    reader= PdfReader(file)
    writer =PdfWriter(file)
    for i in range (1,len(reader.pages)):
        pages = reader.pages[i]
        rotate_page=pages.rotate(90)
        writer.add_page(rotate_page)


    with open('output1.pdf','wb') as file1:
        writer.write(file1)
```

**9. What is the difference between False, True, and None for the bold variable?**


**Ans:**

In the python-docx library, the bold attribute of a Run object is used to determine or set the bold formatting of text. The bold attribute can take three values: True, False, and None. Here's the difference between these values:

**True:**

When bold is set to True, the text in the run is explicitly bold.

**False:**

When bold is set to False, the text in the run is explicitly not bold.

**None:**

When bold is set to None, the text in the run does not have any explicit bold formatting applied.

```python
from docx import Document
doc = Document()
paragraph = doc.add_paragraph("This is a paragraph with ")
run_bold = paragraph.add_run("bold text.")
run_bold.bold = True
run_not_bold = paragraph.add_run(" Not bold text.")
run_not_bold.bold = False
run_inherited = paragraph.add_run(" Inherited bold text.")
run_inherited.bold = None
doc.save('example.docx')
```
