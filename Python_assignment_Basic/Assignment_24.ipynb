**3.Compare and contrast map, filter, and reduce.**

**Ans:**
Map fuction:

 Map is applied on iterable object and return iterator object that can be print into list and tuple format



```python
list1=[1,2,3,4]
saq = (map(lambda x:x*x,list1))
list2=list(saq)
list2
```

filter fuction :

it filter the iterable object and return either True and False

```python
list3=[1,2,3,4,5,6,7,8]
fi = filter(lambda x:(x%2==0),list3)
list4 =list(fi)
list4
```

if we use 'None' so its return only true value

```python
values = [0, 1, 2, "", "Hello", [], [1, 2, 3], None, True, False]
list5 = filter(None , values)
list(list5)
```

The reduce() function, as the name describes, applies a given function to the iterables and returns a single value

```python
from functools import reduce
list6=[1,2,3,4,5,6,7,8]
value= reduce(lambda x,y:x+y,list6)
value
```

map and filter require functions that work on individual elements.

reduce requires a binary function that combines two elements at a time.

**5.What are recursive functions, and how are they used?**


**Ans:**

Recursive functions are functions that call themselves in order to solve a problem. They are used when a problem can be broken down into smaller, simpler subproblems of the same type.

```python
def fact(n):
  if n>1:
    return fact(n-1)*n
  else:
    return 1
fact(5)
```

**6.What are some general design guidelines for coding functions?**


**Ans:**

A function should perform a single task or operation. This makes it easier to understand, test, and maintain.

The function name should clearly describe what the function does.

 Functions should be relatively short. If a function is too long, it might be doing too much and should be broken down into smaller functions.

 A function should take parameters as input and return a result. Avoid using global variables inside functions as much as possible.

  Whenever possible, design functions that always produce the same output for the same input and do not modify external state (pure functions). This makes testing and debugging easier.

  Functions should handle errors gracefully, either by returning a special value (like None or null) or by raising an exception.

  Avoid duplicating code by encapsulating repeated logic in a function. This reduces the chance of errors and makes the codebase easier to maintain.

   Functions should be designed to work independently of other functions as much as possible, facilitating easier testing and maintenance.

   Ensure that functions can be easily tested with unit tests.
