**7.Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to its next() method: 2, 3, 5, 7, 11, ...**

```python
def genPrimes():
    primes = [2]
    yield 2
    n = 3
    while True:
        is_prime = True
        for p in primes:
            if n % p == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(n)
            yield n
        n += 2
```

```python
gen_Primes = genPrimes()
print(next(gen_Primes))
print(next(gen_Primes))
print(next(gen_Primes))
print(next(gen_Primes))
print(next(gen_Primes))
```
