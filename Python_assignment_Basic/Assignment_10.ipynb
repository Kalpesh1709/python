**4.   ZipFile objects have a close() method just like File objects’ close() method. What ZipFile method is equivalent to File objects’ open() method?**





**Ans**

zip file method is equivantent to open function

```python
import zipfile
try:
    with zipfile.ZipFile("hello.zip", mode="a") as archive:
        archive.write("new_hello.txt")
    with zipfile.ZipFile("hello.zip") as archive:
        archive.printdir()
except Exception as e:
    print(e)
```

 **5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf
or .jpg). Copy these files from whatever location they are in to a new folder.****

**Ans:**

```python
import os
import shutil

def sort_and_move_files(source_folder, pdf_folder, image_folder):
  os.makedirs(pdf_folder,exist_ok= True)
  os.makedirs(image_folder,exist_ok= True)
  Image=('.jpg','.jpeg','.png')
  for foldername, subfolders, filenames in os.walk(source_folder):
   for filename in filenames:
     source_file_path = os.path.join(foldername, filename)
     if filename.lower().endswith('.pdf'):
       shutil.move(source_file_path,os.path.join(pdf_folder,filename))
     if filename.lower().endswith(Image):
      shutil.move(source_file_path,os.path.join(image_folder,filename))
```

```python
source_folder = r'C:\Users\thaka\Downloads'
pdf_folder = r'C:\Users\thaka\OneDrive\Desktop\New\pdf_folder'
image_folder = r'C:\Users\thaka\OneDrive\Desktop\New\image_folder'

sort_and_move_files(source_folder, pdf_folder, image_folder)
```
