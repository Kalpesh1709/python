**Q1. Describe the differences between text and binary files in a single paragraph.**



**Ans:**

text files store human-readable characters, while binary files contain data in a binary format, making them unreadable without specialized software. Text files use encodings like ASCII, whereas binary files can store any type of data, such as images or executables, more compactly.

**Q2. What are some scenarios where using text files will be the better option? When would you like to use binary files instead of text files?**



**Ans:**

**Scenarios for Using Text Files:**
- **Configuration Files**: When you need a file that's easy for humans to read and edit, like settings for software.
- **Log Files**: Storing logs for debugging or monitoring, where readability is important.
- **Data Serialization**: Simple data formats like CSV or JSON, which need to be shared or edited across different systems.

**Scenarios for Using Binary Files:**
- **Multimedia Files**: Storing images, videos, or audio, where efficiency and compactness are crucial.
- **Executable Files**: Compiled programs or machine code, which need to be executed by a computer.
- **Large Datasets**: When performance is critical, and you need to store and access large amounts of data quickly, such as in databases or scientific computing.

**Q3. What are some of the issues with using binary operations to read and write a Python integer directly to disc?**



**Ans**

Using binary operations to read and write Python integers can lead to portability issues across systems, risk of data corruption, added complexity in managing byte order and size, lack of readability for debugging, and potential compatibility problems with large integers.

**Q8. When is pickling the best option?**






**Ans:**

Pickling is best used when you need to:

**Save Python Objects:** Store complex Python objects (like lists, dictionaries, or custom classes) to a file so you can reload them later.

**Serialize Data:** Convert Python objects into a format that can be easily saved to disk or sent over a network.

**Preserve State:** Save the state of a Python application or program, such as during debugging or long-running processes.
