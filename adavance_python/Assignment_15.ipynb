**1.What are the new features added in Python 3.8 version?**

**Ans:**

Here are Python 3.8 features with examples:

Walrus Operator (:=): Allows assignment within expressions.

```python
list1= [1,2,3,4,5,7,8,9,9,0,0,0,8,7]
if (n := len(list1)) > 10:
  print(n)
```

Positional-Only Parameters: Use / to specify parameters that must be positional.

```python
def func(a, b, /, c, d):
    return a + b + c + d

func(2,3,6,5)
```

f-Strings with =: Displays both expressions and their values.

```python
x = 10
y = 20
print(f"{x=}, {y=}")  # Output: x=10, y=20
```

TypedDict: Defines dictionaries with fixed key-value types

```python
from typing import TypedDict

class Person(TypedDict):
    name: str
    age: int

person: Person = {"name": "Alice", "age": 30}
```

**2.What is monkey patching in Python?**


**Ans:**
Monkey patching in Python refers to the practice of dynamically modifying or extending classes or modules at runtime. This allows you to change the behavior of code without altering the original source code.

```python
class abc:
  def welcome(self):
    return("hello data")
obj= abc()
print(obj.welcome())
def exitway(self):
  return("exit gate")
abc.welcome = exitway
print(obj.welcome())
```

**3.What is the difference between a shallow copy and deep copy?**


Shallow copy:

its create copy of object it doesnot created copy of nested object for shallow copy **Copy** use

```python
import copy
original=[1,[2,3]]
shallowcopy=copy.copy(original)
shallowcopy[0]=12
shallowcopy[1][0]=23
print(original)
print(shallowcopy)
```

Deep Copy:

A deep copy creates a new object and recursively copies all nested objects. The result is a completely independent copy of the original object

```python
import copy
original=[1,[2,3]]
deep=copy.deepcopy(original)
deep[0]=12
deep[1][0]=23
print(original)
print(shallowcopy)
```

**5.What is generator comprehension?**

**Ans:**


Generator comprehension is a concise way to create a generator in Python. Similar to list comprehensions, it allows you to create generators in a single line of code but uses parentheses () instead of square brackets [].

A generator comprehension produces a generator object, which generates values on the fly and yields one item at a time when iterated over, making it memory-efficient for large datasets.

syntax:

(expression for item in iterable if condition)


```python
gen= (i**2 for i in range(10)if(i%2==0))
```

```python
for j in gen:
  print(j)
```
