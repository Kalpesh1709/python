**Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?**



**Ans:**

A probability distribution describes how the values of a random variable are distributed. It provides the probabilities of different outcomes for that variable.

- **Discrete Probability Distribution**: Lists probabilities for each possible value (e.g., rolling a die).
- **Continuous Probability Distribution**: Describes probabilities over a range of values using a probability density function (e.g., heights of people).

While individual outcomes are random, the distribution provides a model for predicting the likelihood of various outcomes, allowing for statistical predictions and analysis.

**Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?**



**Ans:**

Yes, there’s a distinction:

- **True Random Numbers**: Generated from physical processes, truly unpredictable.
- **Pseudo-Random Numbers**: Generated by algorithms, reproducible with a known seed.

**Why Pseudo-Random Numbers Are “Good Enough”**:
- They mimic true randomness well.
- They are fast and reproducible, suitable for most applications.

**Q6. What kind of object can be shuffled by using random.shuffle?**



**Ans:**

The random.shuffle function can be used to shuffle **mutable sequences** in Python, such as lists.

It cannot be used with immutable sequences like tuples or strings.

**Q9. What are the three logarithmic functions that Python supports?**

**Ans:**

Python supports these three logarithmic functions:
