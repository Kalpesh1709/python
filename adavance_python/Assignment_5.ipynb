**Q1. What is the meaning of multiple inheritance?**



**Ans:**

Multiple inheritance allows a class to inherit from more than one base class. This means the class can use attributes and methods from all its parent classes

```python
class A:
  def class_a(self):
    print("class a")
class B:
  def class_b(self):
    print("class b")
class C(A,B):
  pass
c = C()
c.class_a()
```

**Q2. What is the concept of delegation?**



**Ans:**

Delegation is when an object (the delegator) passes a task to another object  to handle. It allows one object to use the functionality of another without inheriting from it.

```python
class engin:
  def start(self):
    print("engin starting")
class car:
  def __init__(self):
    self.engin = engin()
  def start(self):
    self.engin.start()
Car = car()
Car.start()
```

**Q3. What is the concept of composition?**


**Ans**

Composition is when a class uses objects from other classes as its components. Instead of inheriting from another class, it combines functionalities by including instances of other classes.

```python
class Engine:
    def start(self):
        print("Engine starting")

class Car:
    def __init__(self):
        self.engine = Engine()  # Composition

    def start(self):
        self.engine.start()

car = Car()
car.start()  # Output: Engine starting
```


**Q4. What are bound methodsÂ and how do we use them?**



**Ans:**

Bound methods are methods tied to a class instance. They automatically receive the instance (self) when called.

```python
class Engine:
    def start(self):
        print("Engine starting")
car = Engine()
print(car.start())
```

**Q5. What is the purpose of pseudoprivate attributes?**

**Ans:**

Pseudoprivate attributes in Python are used to prevent accidental access or modification of class attributes from outside the class. They are created by prefixing an attribute name with __ (double underscore)

```python
class my_info:
 def __init__(self):
  self.__value = 54
 def print_data(self):
  return(self.__value)
obj = my_info()
print(obj.print_data())

```
