
**Q4. What is the relationship between indexing and slicing?**



**Ans:**

Indexing is a special case of slicing where only a single element is retrieved. For example, my_list[2] is equivalent to my_list[2:3][0].

Slicing allows for more flexibility by enabling access to multiple elements at once, whereas indexing is limited to just one element.

Both use zero-based indexing: The first element is at index 0.

In summary, while indexing gets you one element, slicing gets you a range of elements. Slicing can be thought of as an extension of indexing.
