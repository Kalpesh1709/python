**1. What is the concept of an abstract superclass?**



**Ans:**

An abstract class/superclass can be considered as a blueprint for other classes. It allows US to create a set of methods that must be created within any child classes built from the abstract class. A class which contains one or more abstract methods is called an abstract class Whereas an abstract method is a method that has a declaration but does not have an implementation



```python
from abc import ABC, abstractmethod
# Super abstract class (very generalized)
class LivingBeing(ABC):
     @abstractmethod
     def breathe(self):
        pass
     @abstractmethod
     def grow(self):
        pass
class Animal(LivingBeing):
    @abstractmethod
    def sound(self):
        pass
class Dog(Animal):
    def sound(self):
        print("Bark!")
    def breathe(self):
        print("Dog is breathing.")
    def grow(self):
        print("Dog is growing.")
dog = Dog()
dog.breathe()
```

**3. Why does a class need to manually call a superclass&#39;s __init__ method?**



**Ans:**

A class needs to manually call a superclass's **__ init __** method to make sure the superclass is properly initialized. This ensures that any setup or attributes defined in the superclass are correctly set up in the subclass.

let see example:


```python
class parent:
  def __init__(self,name):
    self.name = name
    print("parent class init")
class child(parent):
  def __init__(self , name ,age):
    super().__init__(name)
    print("child class init")
    self.age = age
Child = child("rahul",30)
```

**4. How can you augment, instead of completely replacing, an inherited method?**


**Ans:**

inherited method instead of completely replacing it, you can call the superclass's method within the subclass's method. This way, you can add new functionality while still retaining the behavior of the original method

```python
class parent:
  def Thakare(self):
    print("thakare family")
class child(parent):
  def Thakare(self):
    super().Thakare()
    print("kalpesh")
c=child()
c.Thakare()
```


**5. How is the local scope of a class different from that of a function?**

**Ans:**

 Variable which is defined inside a function is local to that function. It is accesible from the point at which it is defined until the end of the function, and exists for as long as the function is existing.

Similary a variable inside of a class also has a local variable scope.
Variables which are defined in the class body (but outside all methods) are called as class level variables or class attributes. They can be referenced by their bare names within the same scope, but they can also be accessed from outside this scope if we use the attribute access operator (.) on a class or an instance of the class.

lets see below example

```python
class Person:
    name = "kalpesh"
    def __init__(self):
        pass
print(Person.name) # Accessing species using class name
Male = Person()
print(Male.name) # Accessing species using instance of class
```
