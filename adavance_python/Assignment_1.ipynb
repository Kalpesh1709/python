**Q2.Where does an inheritance search look for an attribute?**

**Ans:**
lets see below example:

```python
class a:
  def __init__(self):
    self.a_att ='abbtribute A'
class b(a):
  def __init__(self):
    super().__init__()
    self.b_att ='abbtribute B'
class c(b):
  def __init__(self):
    super().__init__()
    self.c_att ='abbtribute c'
obj=c()
print(obj.c_att)
print(obj.b_att)
print(obj.c_att)
```

Explanation of the attribute search:

When accessing **obj.c_att**, the search looks in C first, finds **c_att**, and returns it.

When accessing **obj.b_att**, the search first looks in C, doesn’t find **b_att** there, then looks in B and finds **b_att.**

When accessing **obj.a_att**, the search first looks in C, doesn’t find **a_att** there, then in B, and still doesn’t find it, so it searches in A and finds **a_att**.


**Q3. How do you distinguish between a class object and an instance object?**



**Ans:**

**Class Object:**
 This is the blueprint for creating instances. It is created when you define a class. You can access class-level attributes and methods through the class itself. Class objects represent the structure and behavior that all instances of the class share.

**Instance Object:** This is a specific realization of a class. It is created when you instantiate a class. Each instance object holds its own data and can use the methods defined by the class.

```python
class myclass:
  class_attribute='this is class attribute'
  def __init__(self,value):
    self.instance= value
  def show(self):
    return f'Instance Attribute:{self.instance}'
inst=myclass('hello')
print(myclass.class_attribute)#class attribute
print(inst.show())#instance
```

**Ans**

In a class’s method, the first argument (self) is special because it refers to the instance of the class that the method is being called on. It allows the method to access and modify the instance's attributes and call other methods on that instance.

**Q5. What is the purpose of the __init__ method?**

The __init__ method in Python is called the initializer or constructor method. Its main purposes are:

**Initialize Attributes:** It sets up initial values for the attributes of an instance when the object is created.

**Set Up Object State:** It allows you to configure an object with any required initial state or parameters.
