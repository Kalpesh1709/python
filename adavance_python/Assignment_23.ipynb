**Q1.If you have any, what are your choices for increasing the comparison between different figures on the same graph?**


**Ans:**

To increase comparison between different figures on the same graph, here are concise strategies:

1. **Use Different Colors**: Assign distinct colors to differentiate between figures.
2. **Use Different Line Styles/Markers**: Apply different line patterns (dashed, dotted) or markers (circles, squares).
3. **Add Grid and Axis Labels**: Improve clarity and context for better comparison.
4. **Use Dual Y-Axis**: Plot data with different scales on two Y-axes for better visualization.
5. **Normalize Data**: Scale data to a common range (e.g., percentages) for direct comparison.

These techniques enhance visual contrast and make comparisons more intuitive.

**Q5. Compare and contrast the three types of array multiplication between two numpy arrays: dot product, outer product, and regular multiplication of two numpy arrays**


**Ans:**

**Dot Product:**

Produces a scalar (1D) or matrix (2D) from the sum of products of corresponding elements.

**Outer Product:**

Produces a matrix from multiplying each element of one vector by each element of another.

**Element-wise Multiplication:**

Produces an array of the same shape, with each element being the product of corresponding elements from the input arrays.

**Q6. Before you buy a home, which numpy function will you use to measure your monthly mortgage payment?**



**Ans:**

To measure your monthly mortgage payment before buying a home, you can use the numpy.pmt() function. This function calculates the payment amount for a loan based on constant payments and a constant interest rate.

numpy.pmt(rate, nper, pv, fv=0, when='end')

rate: The interest rate for each period.

nper: The total number of periods (e.g., months).

pv: The present value or principal amount of the loan.

fv: The future value, usually 0 for a mortgage.

when: When payments are due, either 'begin' or 'end' (default is 'end').
