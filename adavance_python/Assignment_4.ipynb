**Q1.Which two operator overloading methods can you use in your classes to support iteration?**



**Ans:**

To support iteration in your classes, you can use the following two operator overloading methods:


__iter__():

This method returns the iterator object itself. It is called when an iteration is initialized, such as in a for loop.

__next__():

This method returns the next item in the sequence. It raises a StopIteration exception when there are no more items to return, signaling the end of the iteration

```python
class my_class:
  def __init__(self,lower,upper):
    self.current = lower
    self.upper = upper
  def __iter__(self):
    return (self)
  def __next__(self):
    if self.current > self.upper:
      raise StopIteration
    else:
      self.current+= 1
      return self.current-1
for ele in my_class(10,30):
    print(ele, end=" ")
```

**Ans:**

**__ str __():**

Used when you print an object or convert it to a string with str(). This method should return a human-readable string representation of the object.


```python
class my_sting:
  def __str__(self):
    return "my name is kalpesh"
name = my_sting()
print(name)
```

Used when you call repr() on an object, or when you're working in the interactive Python shell and the object is returned as an output
It should return an unambiguous string that could be used to recreate the object.

```python
class my_sting:
  def __repr__(self):
    return "my name is kalpesh"
name = my_sting()
print(name)
```

**Q3. In a class, how do you intercept slice operations?**



**Ans:**

In a class, you can intercept slice operations by overloading the **__ getitem __()** and **__ setitem __()** methods. These methods handle indexing and slicing operations.

```python
class MyList:
    def __init__(self, data):
        self.data = data

    def __setitem__(self, key, value):
        self.data[key] = value

my_list = MyList([1, 2, 3, 4, 5])
my_list[1:3] = [20, 30]
print(my_list.data)
```

**Q4. In a class, how do you capture in-place addition?**


**Ans:**

In a class, you can capture in-place addition by overloading the **__ iadd __()**

```python
class caradd:
  def __init__(self,carnumber):
    self.carnumber = carnumber
  def __iadd__(self,total):
    self.carnumber += total.carnumber
    return self.carnumber
b1 = caradd(10)
b2 = caradd(20)
b1+=b2
print(b1)
```

**Q5. When is it appropriate to use operator overloading?**

**Ans:**

Use operator overloading when:

Logical Operators: Your class should support operators like + or ==, similar to built-in types.

Readability: It makes your code more readable.

Custom Behavior: You need specific behavior for operators that fits your classâ€™s purpose.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    def __str__(self):
        return f"result({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)
print(p1 + p2)

```
