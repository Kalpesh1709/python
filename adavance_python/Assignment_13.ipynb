**Q1. Can you create a programme or function that employs both positive and negative indexing? Is there any repercussion if you do so?**



```python
my_list=[2,3,45,65,78,8]

def indexof_list(my_list,position):
  return my_list[position]
print(f"positive indexing:{indexof_list(my_list,5)}")
print(f"negative indexing:{indexof_list(my_list,-2)}")
```

**Q4. Explain the distinctions between indexing and slicing.**



**Ans:**

Indexing returns a single element;
slicing returns a new sequence.

Indexing uses one number;
slicing can use up to three numbers (start, stop, step).

Slicing can be used to create sublists or substrings, while indexing is for accessing individual elements.

**Q6.If you pass a list to a function, and if you want the function to be able to change the values of the list—so that the list is different after the function returns—what action should you avoid?**

**Ans:**

Always use return statement, if we want the see the changes in the input list.

```python
my_list = [1,2,3,4,5,6]
def modify_list(in_list):
    in_list.append(200)
    return in_list
print(modify_list(my_list))
```
