**Q1. What is the concept of a metaclass?**

**Ans:**

A metaclass is a blueprint for creating classes, just like a class is a blueprint for creating objects. Metaclasses allow you to customize how classes are created.

```python
class MyMeta:
  def __new__(cls, name, bases, dct):
    #print(type(MyMeta))
    return super().__new__(cls, name, bases, dct)
    class MyClass(metaclass = Mymeta):
      pass
```

**Q3. How do class decorators overlap with metaclasses for handling classes?**

Class Decorators and Metaclasses both modify classes, but they do so at different stages:

class decoder

1.Applied After class creation.

2.Modifies the class by adding or changing attributes/methods.

lets see below example

```python

def greet(cls):
  cls.gret=lambda self:"hello"
  return cls
@greet
class pqr:
  pass
obj =pqr()
print(obj.gret())
```

Metaclasses

Applied During class creation.

Controls the creation process, allowing deeper modifications.

```python
class MyMeta(type):
  def __new__(cls, name, bases, dct):
    dct['greet']=lambda self:"hello"
    return super().__new__(cls, name, bases, dct)
class MyClass(metaclass = MyMeta):
  pass
obj = MyClass()
print(obj.greet())
```
