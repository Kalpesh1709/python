**Ans:**

Decimal(1.200) and Decimal(1.2) represent the **same numeric value** but have **different internal states** due to their precision.

- **`Decimal(1.200)** has three decimal places.
- **`Decimal(1.2)** has one decimal place.

While they are equal in value (1.2), they are **not identical objects** because their **precision differs**.

**4.Why is it preferable to start a Decimal object with a string rather than a floating-point value?**



**Ans:**

It is preferable to start a Decimal object with a string rather than a floating-point value because floating-point numbers in Python (and most programming languages) are represented using binary approximation. This can introduce small precision errors due to the way floating-point arithmetic works.

When a Decimal object is initialized from a floating-point value, the precision errors of the float are carried over into the Decimal object. On the other hand, when a Decimal object is initialized from a string, it avoids these errors and preserves the exact decimal representation as intended.

**Q9.Consider the following two fraction objects: Fraction(1, 2) and Fraction(1, 2). (5, 10). Is the internal state of these two objects the same? Why do you think that is?**



**Ans:**

Yes, the internal state of Fraction(1, 2) and Fraction(5, 10) is the same. This is because the Fraction class automatically reduces fractions to their simplest form.

Fraction(1, 2) is already in its simplest form.
Fraction(5, 10) simplifies to 1/2.
Python’s Fraction class ensures that all fractions are stored in their reduced form internally, so both objects represent the same value, 1/2.

The Fraction class and the int type have a **containment** relationship. The Fraction class can contain integers as numerators or denominators but does not inherit from int.
