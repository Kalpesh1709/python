**Q1. What is the distinction between a numpy array and a pandas data frame? Is there a way to convert between the two if there is?**



**Ans:**

**NumPy Array:**

**Structure:** Homogeneous (all elements must be of the same data type).

**Functionality:** Primarily used for numerical computations and operations on arrays.

**Indexing:** Uses integer-based indexing.

```python
import numpy as np
np_array = np.array([1, 2, 3])
```

**Pandas DataFrame:**

**Structure:** Heterogeneous (different columns can have different data types).

**Functionality:** Designed for data manipulation and analysis, supporting labeled axes (rows and columns).

**Indexing:** Uses labeled indexing for rows and columns.

```python
import pandas as pd
df = pd.DataFrame(["abc",123])
```

conversion into pandas use **pd.DataFrame()**.

conversion into numpy use **df.to_numpy()**
