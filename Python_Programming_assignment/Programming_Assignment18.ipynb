**Question 1:**
Create a function that takes a list of non-negative integers and strings and return a new list
without the strings.

Examples

filter_list([1, 2, &quot;a&quot;, &quot;b&quot;]) ➞ [1, 2]

filter_list([1, &quot;a&quot;, &quot;b&quot;, 0, 15]) ➞ [1, 0, 15]

filter_list([1, 2, &quot;aasf&quot;, &quot;1&quot;, &quot;123&quot;, 123]) ➞ [1, 2, 123]



```python
def filter_list(l):
  new_list = []
  for i in l:
    if type(i)== int:
       new_list.append(i)
  return new_list
print(f'filter_list([1, 2, "a", "b"]) ➞ {filter_list([1, 2, "a", "b"])}')
print(f'filter_list([1, "a", "b", 0, 15])➞ {filter_list([1, "a", "b", 0, 15])}')
print(f'filter_list([1, 2, "aasf", "1", "123", 123])➞ {filter_list([1, 2, "aasf", "1", "123", 123])}')
```

**Question 2**

The &quot;Reverser&quot; takes a string as input and returns that string in reverse order, with the
opposite case.

Examples

reverse(&quot;Hello World&quot;) ➞ &quot;DLROw OLLEh&quot;

reverse(&quot;ReVeRsE&quot;) ➞ &quot;eSrEvEr&quot;

reverse(&quot;Radar&quot;) ➞ &quot;RADAr&quot;



```python
def reversed(s):
  s = s[::-1]
  #print(s)
  s = s.swapcase()
  return s

print(f'reverse("Hello World") ➞ {reversed("Hello World")}')
print(f'reverse("ReVeRsE") ➞ {reversed("ReVeRsE")}')
print(f'reverse("Radar") ➞ {reversed("Radar")}')
```

first ➞ 1

Your task is to unpack the list writeyourcodehere into three variables, being first,middle, and last, with middle being everything in between the first and last element. Then print all three variables.


**Question 4**
Write a function that calculates the factorial of a number recursively.

Examples

factorial(5) ➞ 120

factorial(3) ➞ 6

factorial(1) ➞ 1

factorial(0) ➞ 1



```python
def fact(n):
  if(n>1):
    n=n*fact(n-1)
    return n
  else:
    return 1
```

```python
print(f'factorial(5)➞ {fact(5)}')
print(f'factorial(3)➞ {fact(3)}')
print(f'factorial(1)➞ {fact(1)}')
print(f'factorial(0)➞ {fact(0)}')
```

**Question 5**
Write a function that moves all elements of one type to the end of the list.

Examples

move_to_end([1, 3, 2, 4, 4, 1], 1) ➞ [3, 2, 4, 4, 1, 1]

 Move all the 1s to the end of the array.

move_to_end([7, 8, 9, 1, 2, 3, 4], 9) ➞ [7, 8, 1, 2, 3, 4, 9]

move_to_end([&quot;a&quot;, &quot;a&quot;, &quot;a&quot;, &quot;b&quot;], &quot;a&quot;) ➞ [&quot;b&quot;, &quot;a&quot;, &quot;a&quot;, &quot;a&quot;]

```python
def move_to_end(l,n):
  for i in l:
    if(i==n):
      l.remove(i)
      l.append(i)
  return l


print(f'move_to_end([1, 3, 2, 4, 4, 1], 1) ➞ {move_to_end([1, 3, 2, 4, 4, 1], 1)}')
print(f'move_to_end([7, 8, 9, 1, 2, 3, 4], 9) ➞{move_to_end([7, 8, 9, 1, 2, 3, 4], 9)}')
print(f'move_to_end(["a", "a", "a", "b"], "a") ➞ {move_to_end(["a", "a", "a", "b"], "a")}')

```
