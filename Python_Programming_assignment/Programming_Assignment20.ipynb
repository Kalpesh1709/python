**Question1**

Create a function that takes a list of strings and integers, and filters out the list so that it returns a list of integers only.

Examples

filter_list([1, 2, 3, &quot;a&quot;, &quot;b&quot;, 4]) ➞ [1, 2, 3, 4]

filter_list([&quot;A&quot;, 0, &quot;Edabit&quot;, 1729, &quot;Python&quot;, &quot;1729&quot;]) ➞ [0, 1729]

filter_list([&quot;Nothing&quot;, &quot;here&quot;]) ➞ []



```python
def filter_list(lst):
  complst=[i for i in lst if type(i)==int]
  return(complst)


print(f'filter_list([1, 2, 3, "a", "b", 4]) ➞{filter_list([1, 2, 3, "a", "b", 4])}')
print(f'filter_list(["A", 0, "Edabit", 1729, "Python", "1729"])➞{filter_list(["A", 0, "Edabit", 1729, "Python", "1729"])}')
print(f'filter_list(["Nothing", "here"]) ➞{filter_list(["Nothing", "here"]) }')
```

**Question2**


Given a list of numbers, create a function which returns the list but with each element&#39;sindex in the list added to itself. This means you add 0 to the number at index 0, add 1 to the number at index 1, etc...

Examples

add_indexes([0, 0, 0, 0, 0]) ➞ [0, 1, 2, 3, 4]

add_indexes([1, 2, 3, 4, 5]) ➞ [1, 3, 5, 7, 9]

add_indexes([5, 4, 3, 2, 1]) ➞ [5, 5, 5, 5, 5]


```python
def add_indexes(lst):
  complst=[i+lst[i] for i in range(len(lst))]
  return complst

print(f'add_indexes([0, 0, 0, 0, 0])➞ {add_indexes([0, 0, 0, 0, 0])}')
print(f'add_indexes([1, 2, 3, 4, 5])➞ {add_indexes([1, 2, 3, 4, 5])}')
print(f'add_indexes([5, 4, 3, 2, 1])➞ {add_indexes([5, 4, 3, 2, 1])}')
```

Create a function that takes the height and radius of a cone as arguments and returns the volume of the cone rounded to the nearest hundredth. See the resources tab for the formula.

Capture1.JPG

Examples

cone_volume(3, 2) ➞ 12.57

cone_volume(15, 6) ➞ 565.49

cone_volume(18, 0) ➞ 0



```python
# cone volume = 1/3Pi r^2 h
from math import pi
from math import pow
def cone_volume(h, r):
  return (1/3)*pi*pow(r,2)*h


print(f'cone_volume(3, 2) ➞ {cone_volume(3, 2):.2f}')
print(f'cone_volume(15, 6) ➞ {cone_volume(15, 6):.2f}')
print(f'cone_volume(18, 0) ➞ {cone_volume(18, 0)}')


```

**Question4**

This Triangular Number Sequence is generated from a pattern of dots that form a triangle.

The first 5 numbers of the sequence, or dots, are:

1, 3, 6, 10, 15

This means that the first triangle has just one dot, the second one has three dots, the third one
has 6 dots and so on.

Write a function that gives the number of dots with its corresponding triangle number of the
sequence.

Examples

triangle(1) ➞ 1

triangle(6) ➞ 21

triangle(215) ➞ 23220



```python
def triangle(d):
  k=0
  for j in range(1,d+1):
    k=j+k
  return k

print(f'triangle(1) ➞  {triangle(1)}')
print(f'triangle(6) ➞  {triangle(6)}')
print(f'triangle(215) ➞  {triangle(215)}')
```

missing_num([7, 2, 3, 6, 5, 9, 1, 4, 8]) ➞ 10
