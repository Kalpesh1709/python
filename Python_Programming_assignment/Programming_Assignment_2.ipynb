**1. Write a Python program to convert kilometers to miles?**

Ans: formula
         miles=km*0.621371

```python
def milesCon(km):
  miles=km*0.621371
  print("%0.2f is km =%0.2f miles"%(km,miles))
```

```python
KM=float(input("enter km"))
milesCon(KM)
```

**2. Write a Python program to convert Celsius to Fahrenheit?**

```python
def fercon(celcius):
  F=float((celcius)*(9/5)+32)
  print("Fahrenheit=%0.2f"%F)
```

```python
C=float(input("input celcius temp"))
fercon(C)
```

**4.Write a Python program to solve quadratic equation?**

```python
print("Equation: ax^2 + bx + c ")
a=int(input("Enter a: "))
b=int(input("Enter b: "))
c=int(input("Enter c: "))
d=b**2-4*a*c
d1=d**0.5
if(d<0):
    print("The roots are imaginary. ")
else:
    r1=(-b+d1)/2*a
    r2=(-b-d1)/2*a
    print("The first root: ",round(r1,2))
    print("The second root: ",round(r2,2))
```
