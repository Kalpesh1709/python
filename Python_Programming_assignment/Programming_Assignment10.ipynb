**6. Write a Python program to find N largest elements from a list?**

```python
def N_largest(list,n):
  list.sort(reverse=True)
  for i in range(n):
    print(list[i])
```

```python
list=[45,672,2673,3748,4738,536,5667,90]
n=int(input("enter the number"))
N_largest(list,n)
```

**7. Write a Python program to print even numbers in a list?**


```python
def even(list):
  print("list input:",list)
  print("even number:")
  for i in list:
    if i%2==0:
      print(i)
```

```python
list=[1,2,34,56,67,7,7]
even(list)
```

**8. Write a Python program to print odd numbers in a List?**


```python
def odd(list):
  print("list input:",list)
  print("odd number:")
  for i in list:
    if i%2!=0:
      print(i)
```

```python
list=[1,2,34,56,67,7,7]
odd(list)
```

**10. Write a Python program to Cloning or Copying a list?**


```python
def cloning(list):
  list1=[]
  for i in list:
    list1.append(i)
  return list1
```

```python
data=[2,3,4,5,6,7,8,8,9]
j=cloning(data)
print("input",data)
print("output",j)
```

**11. Write a Python program to Count occurrences of an element in a list?**

```python
def count_occur(element):
 # print(element)
  count=0
  for i in element:
    count=0
    for j in element:
      if i==j:
        count=count+1
    print("number:",i,"count",count)
```

```python
Number_of_element =[]
j=int(input("how many number of element you want to enter"))
for i in range(1,j+1):
    a=int(input("enter the number"))
    Number_of_element.append(a)
count_occur(Number_of_element)
```
