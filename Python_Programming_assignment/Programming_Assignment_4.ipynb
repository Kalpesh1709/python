**1.Write a Python Program to Find the Factorial of a Number?**

```python
def fact(n):
  if n==0:
    return 1
  else:
    return n*fact(n-1)
```

```python
Number=int(input("enter the number"))
fact(Number)
```

**2.Write a Python Program to Display the multiplication Table?**

```python
def Multi_Table(Num):
  for i in range(1,11):
    print(f"{Num} X {i} ={Num*i}")
```

**4. Write a Python Program to Check Armstrong Number?**

```python
num=int(input("enter the input"))
sum=0
number=num
str_num=str(num)
n=len(str_num)# its measure number of value length
for i in str_num:
  sum = sum+int(i)**(n)# power we are use

if number == sum:
  print("num is armstrong")
else:
  print("num is not armstrong")

```

**5. Write a Python Program to Find Armstrong Number in an Interval?**

```python
def Armstrong_find(num):
  n=str(num)
  sum=0
  number_len=len(n)
  for i in n:
    sum=sum+int(i)**number_len
  if(num==sum):
    print(f"{num}is armstrong")
```

```python
start_int =int(input("enter starting interval"))
end_int =int(input("enter ending interval"))
for j in range(start_int,end_int):
  Armstrong_find(j)
```
