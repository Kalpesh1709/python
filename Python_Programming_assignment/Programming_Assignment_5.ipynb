**1. Write a Python Program to Find LCM?**

Ans:

The LCM of two integers a and b is denoted by LCM (a,b). It is the smallest positive integer that is divisible by both "a" and "b".

```python
def LCM_number(num1,num2):
    
    #check which number is greater than each other
    if (num1>num2):
        maxnum= num1
    else:
        maxnum =num2
    while(True):
       if(maxnum%num1==0) and (maxnum%num2==0):
        lcm=maxnum
        break
       maxnum+=1
    return lcm

  
```

```python
LCM_number(2,4)
```

**2. Write a Python Program to Find HCF?**

```python
def HCF_number(num1,num2):
    #check which number is less than each other
    if (num1>num2):
        min_num= num2
    else:
        min_num =num1
    for i in range(1,min_num+1):
        if(num1%i==0) and (num2%i==0):
            hcf=i
    return hcf  
   
```
