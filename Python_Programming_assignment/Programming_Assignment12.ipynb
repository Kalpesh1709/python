**2. Write a Python program to find the sum of all items in a dictionary?**


```python
def dicti_add(dicti):
  list=[]
  for i in dicti.values():
    list.append(i)
  return sum(list)

```

```python
dicti={'a':20,'b':30,'c':40}
print(dicti_add(dicti))
```

**5. Write a Python program to insertion at the beginning in OrderedDict?**



```python
from collections import OrderedDict
Test_dict={'days':['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'],
      'num':[1,2,3,4,5,6,7]}
orderupdate=OrderedDict(Test_dict)
orderupdate.update({'month':['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']})
orderupdate.move_to_end('month',last=False)
print(orderupdate)
```

**6. Write a Python program to check order of character in string using OrderedDict()?**

```python
from collections import OrderedDict
def checkOrder(string, pattern):
    dic = OrderedDict.fromkeys(string)
    ptr = 0
    for key,value in dic.items():
        if (key == pattern[ptr]):
            ptr = ptr + 1
        if (ptr == (len(pattern))):
            return 'True'
    return 'False'

string = 'Kalpesh da'
pattern = 'eh'
print (checkOrder(string,pattern))

string2= 'hello'
pattern2= 'llo'
print (checkOrder(string2,pattern2))
```
