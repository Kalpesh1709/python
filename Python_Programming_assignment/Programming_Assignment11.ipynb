**1. Write a Python program to find words which are greater than given length k?**


```python
def find_greater_than(data,length):
  data=data.split()
  for i in data:
    if len(i)>length:
      print(i)
```

```python
data=input("enter the string")
length=int(input("enter string lenth"))
find_greater_than(data,length)
```

**2. Write a Python program for removing i-th character from a string?**


```python
def remove_char(data,i):
  data=data[:i-1]+data[i+1:]
  return data
```

```python
data_1=input("enter the string")
i=int(input("enter the index"))
remove_char(data_1,i)
```

**4. Write a Python to check if a given string is binary string or not?**

```python
def check_binary(data):
  for i in data:
    if i!="0" and i!="1":
      print("not binary")
      break
  else:
    print("binary")
```

```python
data=input("enter input")
check_binary(data)
```

**5. Write a Python program to find uncommon words from two Strings?**


```python
def uncommon_words(data1,data2):
  string1=[]
  string2=[]
  string1=data1.split()
  string2=data2.split()
  for i in string1:
    if i not in string2:
      print("not in String2",i)
  for j in string2:
    if j not in string1:
      print("not in String1",j)
```

```python
string1=input("enter string1")
string2=input("enter string2")
uncommon_words(string1,string2)
```


**6. Write a Python to find all duplicate characters in string?**

```python
def duplicate_char(data):
  for i in data:
    if data.count(i)>1:
      print(i)
```

```python
duplicate_char(" hello world")
```

**7. Write a Python Program to check if a string contains any special character?**

```python
s = '[@_!#$%^&*()<>?/\|}{~:]'
data=input("enter the string")
for i in data:
  if i in s:
    print("contains special character")
    break
else:
  print("not contains special character")
```
